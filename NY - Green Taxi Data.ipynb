## NYC Green Taxi Data Analysis (Sept' 15)

```python
# import essential python packages
import numpy as np
import pandas as pd

import os, json, requests, pickle
from urllib.request import urlretrieve

from math import sqrt, sin, cos, atan2, pi
from haversine import haversine
from datetime import datetime
from calendar import timegm

import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline

from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

from scipy import stats

import warnings
warnings.filterwarnings('ignore')

# define preprocessing functions and other helper functions

# compute proximity of pickup and dropoff points with airport and label if within threshold value (< 1 mile).
def isNearAirport(row):
    #Pickup, Dropoff
    points = [(round(row['Pickup_latitude'], 4) , round(row['Pickup_longitude'], 4)), (round(row['Dropoff_latitude'], 4), round(row['Dropoff_longitude'], 4))]
    #JFK, EWR, LGA lat lon pairs
    airports = [(40.6442, -73.7822), (40.6897, -74.1745), (40.7747, -73.8719)]
    labels = []
    
    # Compute distance of each point from the list of airports and identify if its less than 1 mile.
    # Label this point with index of corresponding airport
    # 0 - Not an airport, 1 - JFK, 2 - EWR, 3 - LGA
    for pt in points:
        distances = []
        for apt in airports:
            distances.append(haversine(pt, apt, miles=True))
        min_dist = np.min(distances) 
        label    = distances.index(min_dist)
        labels.append((label+1) if (min_dist < 1) else 0)

    return labels

# convert provided value with given string representation into datetime object
def str_to_DateTime(val):
    return datetime.strptime(val, '%Y-%m-%d %H:%M:%S')

# extact pickup-hour, pickup-day of the month (1-30), trip time in hours and average speed in miles/hour for given row.
def dist_features(row):
    labels = isNearAirport(row)
    
    d1 = str_to_DateTime(row['lpep_pickup_datetime'])
    d2 = str_to_DateTime(row['Lpep_dropoff_datetime'])
    Pickup_hr = d1.hour
    Pickup_dom = d1.day
    Trip_time = (((d2 - d1).total_seconds())/3600)
    Average_speed = (0 if(Trip_time == 0) else (row['Trip_distance']/Trip_time))
    
    return pd.Series({'c1': labels[0], 
                      'c2': labels[1],
                      'c3': Pickup_hr,
                      'c4': Pickup_dom,
                      'c5': Average_speed
                     }).astype('int')

# extact pickup-hour, dropoff-hour, pickup-day, dropoff-day, trip time in hours for given row.
def process(row):
    labels = isNearAirport(row)
    d1 = str_to_DateTime(row['lpep_pickup_datetime'])
    d2 = str_to_DateTime(row['Lpep_dropoff_datetime'])
    
    Pickup_hr   = d1.hour
    Dropoff_hr  = d2.hour
    Pickup_day  = d1.weekday()
    Dropoff_day = d2.weekday()
    Trip_time   = ((d2 - d1).total_seconds())/3600

    return pd.Series({'c1': labels[0], 
                      'c2': labels[1],
                      'c3': Pickup_hr,
                      'c4': Dropoff_hr,
                      'c5': Pickup_day,
                      'c6': Dropoff_day,
                      'c7': Trip_time
                     }).astype('int')

# Derive different binned and transformed features from date
def time_features(dt):
    dt = str_to_DateTime(dt)
    #bins=96 - indicates the day has been divided into 15 min slots
    minutes_per_bin = int((24 / float(96)) * 60)   
    num_minutes = ((dt.hour * 60) + dt.minute)
    time_bin = num_minutes / minutes_per_bin
    hour_bin = num_minutes / 60
    min_bin = (time_bin * minutes_per_bin) % 60
    time_num = (((hour_bin * 60) + min_bin + minutes_per_bin) / 2.0) / (60 * 24)
    
    #https://www.kaggle.com/zeroblue/mad-scripts-battle-z/comments
    time_cos = cos(time_num * 2 * pi)
    
    return pd.Series({'c8': int(time_bin), 
                      'c9': time_num, 
                      'c10': time_cos
                     })
```

#### Feature Engineering
1. Trip cost - negative values: 
We can clearly observe the features 'Fare_amount', 'Total_amount' and 'taxes' having negative values. This is the result of erronous measurement as trip cost and taxes can not be negative. Also no furthur detail about these measurements is provided by the NYC-TLC. We proceed by removing these transactions. Minimum fare amount must be greater than $2.5 as per the NYC Taxi & Limousine Commission.
2. Non credit transactions:
Credit transactions denoted by 'Payment_type=1' have 'Tip_amount' populated automatically while other transactions have the tip amounts marked 0. All cash tips have also been recorded as 0. As a result, these transactions can not be used for tip percentage prediction.
3. Missing/zero values (latitude-longitude):
No missing or zero values persist in data for latitude or longitude after removal of transactions with negative trip-costs.
4. Other discarded features:
Ehail fee has large number of missing values and hence the feature has been discared. Also Store_and_fwd_flag has no predictive value as it is a flag related to te way trip detail were logged and is not related to the problem at hand.

#### Removing extreme outliers
1. Trip distance:
Records with trip distances larger than 100 miles have been discarded.

#### Feature Creation & Decomposition
New features have been created through decomposition of the 'Pickup' and 'Dropoff' date-time object. 
1.  Pickup_day: weekday of pick-up
2.  Dropoff_day: weekday of dropoff
3.  Pickup_hr: hour of pick-up
4.  Dropoff_hr: hour of dropoff
5.  Fare_amount: modified field generated from difference of Total_amount and Tip_amount for non-zero tip and total fare.
6.  Dropoff_is_weekday: flag created from Dropoff_day to represent if trip was on a weekday or a weekend
7.  Trip_time: Difference between trip pickup and dropoff time in hours.
8.  Average_speed: Trip_distance by Trip_time (miles/hr)
9.  Pickup_pt: indicator if trip pickup was at airport or elsewhere.
10. Dropoff_pt: indicator if trip dropoff was at airport or elsewhere.
11. D_Time_bin: time of the day as a categorical feature.
12. D_Time_num: Time of the day binned to a floating point number between 0 to 1 taking into account the actual time passed, given the bins defined.(If the day is divided into 96 bins, we get time slots in increment of 15 minutes.)
13. D_Time_cos: The D_Time_num is subjected to a cosine transformation. If we take a sine transformation, 11am and 1pm are close to each other, but mid-day and midnight end up having the same value which is harmful for the model. However, if we use the cosine transformation, mid-day and midnight are separated by maximum value possible. 

#### Evaluating feature importance
For Identifying and dropping features with low predctive power, an Extra Trees regressor ensemble model is used to identify feaures which help best predict the response. The model provides importance scores to all the features used in prediction which can be used for feature selection.

```python
tips_df = df[(df['Payment_type'] == 1) &
             (df['Total_amount'] > 0) & 
             (df['Tip_amount'] > 0)]

tips_df['Fare_amount'] = tips_df['Total_amount'] - tips_df['Tip_amount']
tips_df  = tips_df[tips_df['Fare_amount'] > 0]
tips_df['Perc_tip'] = (tips_df['Tip_amount']/tips_df['Fare_amount']) * 100

cols = ['Pickup_pt', 'Dropoff_pt', 'Pickup_hr', 'Dropoff_hr', 'Pickup_day', 'Dropoff_day', 'Trip_time']
tips_df[cols] = tips_df.apply(lambda row: (process(row)), axis=1)

cols = ['D_Time_bin', 'D_Time_num', 'D_Time_cos']
tips_df[cols] = tips_df.apply(lambda row: (time_features(row['Lpep_dropoff_datetime'])), axis=1)

tips_df['VendorID'] = tips_df['VendorID'].map({1:0, 2:1})
tips_df['Dropoff_is_weekday'] = (tips_df['Dropoff_day']/5).astype('int')
tips_df['Average_speed'] = tips_df['Trip_distance'] / tips_df['Trip_time']
tips_df['Average_speed'] = tips_df['Average_speed'].replace([np.inf, -np.inf], 0)
tips_df['Average_speed'] = tips_df['Average_speed'].fillna(0)

tips_df.drop(['lpep_pickup_datetime',
              'Lpep_dropoff_datetime',
              'Payment_type',
              'Total_amount',
              'Tip_amount',
              'Ehail_fee',
              'Store_and_fwd_flag'], 
             inplace=True, axis=1)
```

```python
Y = tips_df['Perc_tip']
X = tips_df.drop(['Perc_tip'], axis=1)

forest = ExtraTreesRegressor(n_estimators=50, random_state=1000)
forest.fit(X, Y)
importances = forest.feature_importances_
indices = np.argsort(importances)[::-1]

l = []
labels = X.columns
for i in indices:
    l.append(labels[i]) 

plt.figure(figsize=(9, 4))
plt.style.use('seaborn-whitegrid')
plt.bar(range(X.shape[1]), importances[indices], align="center", color='black')
plt.xticks(range(X.shape[1]), l, rotation='45')
plt.xlim([-1, X.shape[1]])
plt.show()
```

We clearly observe some of the features to have almost negligible predictive power/contributions. We can go ahead and drop the features with scores below a preset threshold. I have not chosen to do so as I will be leveraging a ensemble based model which is a very robust non-linear machine learning technique. These features may be a problem for simple linear models in which case we may eliminate such features/variables.

We now proceed with application of Random forest and Gradient Boosting Regressors which are ensemble models based on bagging and boosting techniques respectively. They tent to be invarient to noise, and perform better in most cases of data modelling. They will not work for all and every modelling challenge but in general score consistently well in most of the scenarios.
Also being non-linear tree based models feature scaling is not required.

We employ 'Grid Search' for hyper-parameter optimization and use 'Stratified KFold' to sample training and test sets so as to reduce chances of overfitting through the use of hold out prediction set.

```python
tips_df = tips_df[(tips_df['Trip_distance'] > 0) & (tips_df['Trip_distance'] < 100)]

Y = tips_df['Perc_tip']
X = tips_df.drop(['Perc_tip'], axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

model = {
    'rfr': RandomForestRegressor(),
    'gbr': GradientBoostingRegressor()
}

model_parameters = {
        'rfr':{
            'n_estimators': [50, 100, 150], 
            'max_depth': [None, 1, 2, 3, 5], 
            'min_samples_split': [2, 3, 5]
        },
        'gbr':{
            'loss':['lad', 'huber','ls'],
            'n_estimators': [50, 100, 150], 
            'criterion':['friedman_mse','mae'],
            'max_depth': [None, 1, 2, 3, 5], 
            'min_samples_split': [2, 3, 5]
        }
}

skf = StratifiedKFold(n_splits=5)
for m in model:
    clf = GridSearchCV(estimator = model[m], param_grid = model_parameters[m], cv=skf, scoring='neg_mean_squared_error')
    clf.fit(X_train, Y_train)
    print(clf.best_params_)

    Y_pred = clf.predict(X_test)
    residuals = (Y_pred - Y_test)

    print("Grid scores: ", clf.grid_scores_)
    print("Best score on training set: ", clf.best_score_)

    r2 = r2_score(Y_test, Y_pred)
    print("R2 score for model:", m, " is: ", r2)

    mse_metric = mean_squared_error(Y_test, Y_pred)
    print("MSE score for model:", m, " is: ", mse_metric)

    print('Writing best estimator to file for: ', m)
    fname = m + '-model.pkl'
    with open(fname,'wb') as f:
        pickle.dump(clf.best_estimator_, f)
        f.close()
```

### Determine if the average trip speeds are materially the same in all weeks of September using Kruskals test or F-test

```python
dist_df['Pickup_wom'] = np.ceil(dist_df['Pickup_dom']/7).astype('int')
speed_weeks = dist_df[['Pickup_wom', 'Average_speed']].groupby(['Pickup_wom', 'Average_speed']).size()
speed_weeks = np.log(speed_weeks)
```

```python
F, fp = stats.f_oneway(speed_weeks[1], speed_weeks[2], speed_weeks[3], speed_weeks[4])
print('F-test \n', 'statistic:', F, 'p-value:', fp)

K, kp = stats.kruskal(speed_weeks[1], speed_weeks[2], speed_weeks[3], speed_weeks[4])
print('Kruskal-test \n', 'statistic:', F, 'p-value:', kp)
```

The distribution for the Average speed aggregated for each week represents one sample. In total if we divide days in September by 7 we get 5 weeks, with the 5th week having only two days. This week is discarded for the purpose of the test.
Also the aggregated samples do not exibit a normal distribution, log transformation manage to reduce the inherent skew to some extent. Both One-way ANOVA and non-parametric KruskalWillis test is used to evaluate if the weeks display deviation.

As ANOVA assumes normality of underlying distribution, this test can not be trusted but is used only for reference.
The Kruskal-Wallis test is on the other hand a non-parametric test for comparing multiple sample distributions (based on median) and does not assume equal sample size nor normality and hence is better suited for the current scenario.

Both the tests generate very high p-values depicting that the distributions of the 4 week samples are quite similar.
